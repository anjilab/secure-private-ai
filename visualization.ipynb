```python
import matplotlib.pyplot as plt
import random
import PIL 
import os
```

```python
random.seed(42)
```

# FGSM-Adv image visualizaton against epsilon. 
For this code, to work, you must have adv_imgs directory.

```python
epsilons = [0.01, 0.03, 0.1, 0.5]
# plot_selected_files =  os.listdir(os.path.join(cfg["data_root"], 'kvasir_polyp/images')) # 1000
plot_selected_files_kvasir = os.listdir(os.path.join('adv_imgs/fgsm/clipseg_dense_adapter_vl_kvasir_polyp', "0.01")) # 100
random_files_kvasir = random.sample(plot_selected_files_kvasir, 5)


fig, axs = plt.subplots(len(random_files_kvasir), 5  , figsize=(5,5))
for idx, filename in enumerate(random_files_kvasir):
    img_list = [os.path.join('adv_imgs/fgsm/clipseg_dense_adapter_vl_kvasir_polyp', f"{epsilon}", filename) for epsilon in epsilons]
    img_list.insert(0, os.path.join('data/kvasir_polyp/images', filename))
    # img_list.append(os.path.join('data/original_imgs/images', filename))
    
    for col, img_path in enumerate(img_list):
    # Load the image
        if os.path.exists(img_path):
            img = PIL.Image.open(img_path)
        else:
            img = None  # Handle missing files (optional)
            print(f"File not found: {img_path}")
            

        # Plot the image
        if img is not None:
            axs[idx, col].imshow(img)
            axs[idx, col].axis('off')  # Hide axes
            if idx == 0:
                if col == 0:
                    axs[idx, col].set_title(f'Original img')
                else:
                    axs[idx, col].set_title(f"={epsilons[col-1]}")
        else:
            axs[idx, col].axis('off')  # Keep the layout consistent

# Adjust layout and show the plot
# plt.subplots_adjust(wspace=0.8, hspace=0.1)
plt.tight_layout()
plt.show()
```
